```python
# these can be installed with
#   pip install tskit msprime stdpopsim
import tskit, stdpopsim
from IPython.display import SVG, display, clear_output, Image
import time
import numpy as np
```

```python
species = stdpopsim.get_species("HomSap")
contig = species.get_contig("chr22")
model = species.get_demographic_model('OutOfAfrica_3G09')
samples = model.get_samples(3, 3, 3)
engine = stdpopsim.get_engine('msprime')
```

```python
ts = engine.simulate(model, contig, samples, random_seed=42)
```

```python
import msprime
small_ts = msprime.simulate(10, length=2e3, recombination_rate=1e-8, Ne=1e4)
```

```python
SVG(small_ts.draw_svg(size=(800,300)))
```

```python
sub_ts = ts.keep_intervals([[0.0, 2000.0]]).trim()
```

```python
SVG(sub_ts.draw_svg(size=(800,300), force_root_branch=True))
```

```python
sub_ts.draw_svg("example.svg", size=(800,300), force_root_branch=True);
```
